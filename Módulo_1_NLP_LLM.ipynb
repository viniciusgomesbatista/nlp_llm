## Stopwords

“Essas palavras ajudam a diferenciar textos?”

Conceito:

- Palavras muito frequentes
- Pouca carga semântica
- Exemplos: de, o, a, e, para, com

```python
texto
```

```python
# importação de bibliotecas
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stopwords_pt = stopwords.words('portuguese')

stopwords_pt
```

```python
tokens = texto.lower().split()
tokens
```

```python
tokens_sem_stopwords = [t for t in tokens if t not in stopwords_pt]
tokens_sem_stopwords
```

Alerta importante:
- Stopwords dependem do contexto
- Às vezes removê-las piora o modelo

## Vocabulário

“Tudo que não estiver no vocabulário é invisível para o modelo.”

Conceito
- Conjunto de tokens únicos de um corpus
- Define o “universo” de palavras conhecidas pelo modelo

```python
textos = [
    "eu gosto de dados",
    "dados são importantes",
    "ciência de dados é incrível"
]

tokens = []

for texto in textos:
    tokens.extend(texto.split())

vocabulario = set(tokens)
vocabulario
```

# Frequência

Conceito
- Quantas vezes cada palavra aparece
- Base de quase todas as técnicas iniciais de NLP
- Palavras frequentes ≠ palavras importantes

```python
tokens
```

```python
from collections import Counter

frequencia = Counter(tokens)
frequencia
```

## Bag of Words (BoW)

“O modelo só sabe se a palavra aparece, não onde.”

Conceito
- Representa texto como contagem de palavras
- Ignora ordem
- Cada texto vira um vetor numérico

```python
textos
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(textos)
bag_of_words
```

```python
vectorizer.get_feature_names_out()
```

```python
bag_of_words.toarray()
```

# N-grams

Conceito
- Sequências de N tokens
- Capturam um pouco de contexto
- Ex: “ciência de dados” ≠ “ciência” + “dados”

```python
textos
```

```python
vectorizer = CountVectorizer(ngram_range=(2, 2))

ng = vectorizer.fit_transform(textos)

vectorizer.get_feature_names_out()
```

# TF-IDF (intuição)

“TF-IDF diminui o peso de palavras comuns e valoriza palavras distintivas.”

Conceito (sem matemática)

TF:
- palavra frequente no documento
- IDF: palavra rara no conjunto de documentos
- Destaque palavras importantes para aquele texto

```python
textos
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(textos)

X.toarray().round(3)
```

```python
vectorizer.get_feature_names_out()
```

# Tratamento de textos

```python
texto = "Os   Professores foram muito atenciosos! As aulas estavam EXCELENTES, mas o material poderia ser melhor."
```

```python
# Função de limpeza simples (PT-BR)
import re

def limpar_texto(texto):
  texto = texto.lower() #Deixar tudo minúsculo (lowercase)
  texto = re.sub(r'http\S+', '', texto) # remove URLs
  texto = re.sub(r'[^a-zà-ú\s]', '', texto)   # remove pontuação/números
  texto = re.sub(r'\s+', ' ', texto).strip()  # espaços extras
  return texto
```

```python
texto_limpo =limpar_texto(texto)
print(texto_limpo)
```

# Normalização

No nosso caso:
- lower case
- remoção de acentos ❌ (opcional, vamos manter)

Diga explicitamente:
- Não existe “normalização perfeita”, existe adequada ao problema.

```python
import unicodedata

def normalizar(texto):
    return unicodedata.normalize("NFKD", texto).encode("ASCII", "ignore").decode("utf-8")
```

```python
texto_normal = normalizar(texto_limpo)
print(texto_normal)
```

# Lematização

Lematizar = reduzir palavras à sua forma base

Exemplo:
- professores - professor
- aulas - aula
- rapidamente - rápido

```python
nltk.download("rslp")
from nltk.stem import RSLPStemmer

stemmer = RSLPStemmer()
```

```python
tokens_lematizados = [
    stemmer.stem(t) for t in texto_normal.split()
]

print(tokens_lematizados)
```

# Introdução a Large Language Models (LLMs)

## 1. O que é um Transformer

O **Transformer** é uma arquitetura de rede neural criada para processar sequências de dados (principalmente texto).

Ele foi apresentado no paper **Attention Is All You Need**, publicado em 2017 por pesquisadores do **Google Research**.

Antes do Transformer, modelos de linguagem usavam principalmente:

* **RNN (Recurrent Neural Networks)**
* **LSTM (Long Short-Term Memory)**

Esses modelos tinham problemas para lidar com dependências longas no texto e eram difíceis de paralelizar.

O Transformer resolveu isso utilizando um mecanismo chamado:

### Attention

A **attention** permite que o modelo observe todas as palavras de uma frase ao mesmo tempo e determine quais são mais relevantes.

Exemplo:

Frase:

```
O aluno que estudou muito passou na prova.
```

Para entender a palavra **"passou"**, o modelo precisa relacionar com **"aluno"**.

O mecanismo de **self-attention** calcula essas relações entre todas as palavras.

---

### Estrutura simplificada do Transformer

O modelo é composto por blocos repetidos:

```
Entrada (tokens)
      ↓
Embedding
      ↓
Self-Attention
      ↓
Feed Forward Neural Network
      ↓
Normalização
      ↓
Saída
```

Esses blocos são empilhados muitas vezes (12, 24, 48 ou mais camadas).

Arquiteturas modernas como **OpenAI** **GPT**, **Meta AI** **LLaMA**, e **Google DeepMind** **Gemini** são baseadas nessa arquitetura.


# 2. O que é um Large Language Model (LLM)

Um **Large Language Model (LLM)** é um modelo de inteligência artificial treinado para entender e gerar texto.

Ele aprende padrões da linguagem analisando **grandes volumes de dados textuais**.

O treinamento geralmente inclui:

* livros
* artigos
* páginas da web
* código
* fóruns
* documentos

O objetivo do modelo é aprender a **prever a próxima palavra de uma sequência**.

Exemplo:

Entrada:

```
O Brasil fica na América
```

O modelo prevê que a próxima palavra provavelmente é:

```
Latina
```

Esse processo é chamado de **modelagem de linguagem**.

---

# 3. O conceito de Tokens

Modelos de linguagem não trabalham diretamente com palavras.

Eles trabalham com **tokens**.

Tokens são pequenas unidades de texto que podem ser:

* palavras
* partes de palavras
* caracteres
* pontuação

Exemplo de tokenização:

Frase:

```
Inteligência artificial está mudando o mundo
```

Possível tokenização:

```
[Inteligência]
[artificial]
[está]
[mudando]
[o]
[mundo]
```

Ou:

```
[Inte]
[ligência]
[artificial]
[est]
[á]
[mundo]
```

Isso depende do algoritmo de tokenização usado.

Alguns tokenizers comuns:

* BPE (Byte Pair Encoding)
* WordPiece
* SentencePiece

Em modelos como **GPT-4**, um token costuma representar aproximadamente:

```
1 token ≈ 3 a 4 caracteres
```

Ou aproximadamente:

```
1 token ≈ 0.75 palavras
```

---

# 4. Probabilidade da próxima palavra

LLMs funcionam essencialmente como **modelos probabilísticos de linguagem**.

Eles calculam:

```
P(próxima_palavra | contexto)
```

Ou seja:

```
Qual a probabilidade da próxima palavra dado o texto anterior?
```

Exemplo:

Entrada:

```
Hoje o dia está muito
```

O modelo pode gerar probabilidades como:

| Palavra | Probabilidade |
| ------- | ------------- |
| quente  | 0.45          |
| frio    | 0.25          |
| bonito  | 0.10          |
| chuvoso | 0.08          |

O modelo escolhe a próxima palavra com base nessas probabilidades.

Esse processo se repete várias vezes até gerar uma frase completa.

---

# 5. Inferência

**Inferência** é o processo de **usar um modelo treinado para gerar respostas**.

Existem duas fases principais em um LLM:

### Treinamento

O modelo aprende padrões da linguagem a partir de dados.

Essa fase é extremamente custosa.

Pode envolver:

* bilhões ou trilhões de tokens
* milhares de GPUs
* semanas ou meses de processamento

---

### Inferência

Depois do treinamento, o modelo pode ser usado para gerar texto.

Fluxo de inferência:

```
Prompt do usuário
        ↓
Tokenização
        ↓
Passagem pelo modelo
        ↓
Cálculo de probabilidades
        ↓
Escolha do próximo token
        ↓
Decodificação em texto
```

Esse processo acontece **token por token**.

---

# 6. Prompts

Um **prompt** é simplesmente o texto que enviamos para o modelo.

Exemplo:

```
Explique o que é aprendizado de máquina em termos simples.
```

O modelo utiliza esse texto como **contexto inicial** para gerar a resposta.

O design de prompts é chamado de:

**Prompt Engineering**

Algumas estratégias comuns:

### Zero-shot

Sem exemplos.

```
Traduza para inglês:

Eu gosto de estudar.
```

---

### Few-shot

Fornecendo exemplos.

```
Português: casa
Inglês: house

Português: livro
Inglês: book

Português: carro
Inglês:
```

---

### Chain-of-thought

Forçando o modelo a explicar o raciocínio.

```
Resolva passo a passo:
Se João tem 5 maçãs e come 2, quantas restam?
```

---

# 7. Resumo conceitual

Podemos resumir o funcionamento de um LLM da seguinte forma:

```
Texto
  ↓
Tokenização
  ↓
Transformer
  ↓
Cálculo de probabilidades
  ↓
Geração do próximo token
  ↓
Texto gerado
```

Ou em forma simples:

```
LLM = Transformer + Muito Dados + Treinamento massivo
```

---

# 8. Intuição final

Uma forma simples de entender LLMs:

> Um LLM é essencialmente um sistema que **aprendeu padrões da linguagem humana e prevê a próxima palavra com base no contexto**.

Apesar da simplicidade da ideia, a escala dos modelos modernos permite realizar tarefas complexas como:

* responder perguntas
* escrever código
* resumir textos
* traduzir idiomas
* analisar documentos
* interagir em linguagem natural

## Criando um mini chat sobre o meu arquivo

````python
from google.colab import ai

# Contexto fixo do documento
contexto_documento = texto_completo

# Histórico da conversa
historico = []

print("Chat iniciado. Digite 'sair' para encerrar.\n")

while True:
    pergunta = input("Você: ")

    if pergunta.lower() == "sair":
        print("Encerrando chat.")
        break

    # Construção do prompt com contexto + histórico
    prompt = f"""
Você é um assistente que responde perguntas exclusivamente com base no documento abaixo.

Documento:
{contexto_documento}

Histórico da conversa:
{historico}

Pergunta atual:
{pergunta}

Responda de forma objetiva e baseada apenas no documento.
"""

    resposta = ai.generate_text(prompt)

    print("\nAssistente:", resposta, "\n")

    historico.append(f"Pergunta: {pergunta}")
    historico.append(f"Resposta: {resposta}")
````
